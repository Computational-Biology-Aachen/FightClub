```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Define the fitness functions for the first and second models
def fitness_first_model(x, params):
    x_C, x_D, x_P = x
    r_c, alpha, beta, nu, r_p, eta = params
    
    f_C = r_c * x_C + (r_c - alpha) * x_D + (r_c - beta) * x_P
    f_D = alpha * x_C - nu * x_D
    f_P = (r_p - eta) * x_C + r_p * x_P
    
    return f_C, f_D, f_P

def fitness_second_model(x, params):
    x_C, x_D, x_P = x
    r_c, alpha, beta, nu, r_p, eta, lambda_, gamma = params
    
    f_C = r_c - lambda_ * x_C + (r_c - alpha) * x_D + (r_c - beta) * x_P
    f_D = alpha * x_C - nu * x_D
    f_P = (r_p - eta) * x_C + r_p - gamma * x_P
    
    return f_C, f_D, f_P

# Define the replicator dynamics equations for both models
def replicator_dynamics(x, t, params, model='first'):
    x_C, x_D, x_P = x
    if model == 'first':
        f_C, f_D, f_P = fitness_first_model(x, params)
    elif model == 'second':
        f_C, f_D, f_P = fitness_second_model(x, params)
    
    avg_fitness = x_C * f_C + x_D * f_D + x_P * f_P
    dx_C = x_C * (f_C - avg_fitness)
    dx_D = x_D * (f_D - avg_fitness)
    dx_P = x_P * (f_P - avg_fitness)
    
    return [dx_C, dx_D, dx_P]

# Initial parameters for both models
params_first_model = [1.0, 0.5, 0.2, 0.1, 1.0, 0.3]  # r_c, alpha, beta, nu, r_p, eta
params_second_model = [1.0, 0.5, 0.2, 0.1, 1.0, 0.3, 0.4, 0.4]  # r_c, alpha, beta, nu, r_p, eta, lambda, gamma

# Time range for simulation
t = np.linspace(0, 50, 500)

# Function to plot phase diagram on the simplex
def plot_phase_diagram(params, model_name):
    fig, ax = plt.subplots(figsize=(8, 8))
    
    # Plot vector field
    Y, X = np.mgrid[0:1:20j, 0:1:20j]
    U = np.zeros_like(X)
    V = np.zeros_like(Y)
    W = np.zeros_like(Y)
    
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            x_C = X[i, j]
            x_D = Y[i, j]
            x_P = 1 - x_C - x_D
            if x_P >= 0:
                dx = replicator_dynamics([x_C, x_D, x_P], 0, params, model=model_name)
                U[i, j] = dx[0]
                V[i, j] = dx[1]
    
    # Normalize vectors
    norm = np.sqrt(U**2 + V**2)
    U, V = U / norm, V / norm
    
    ax.quiver(X, Y, U, V, color='r')
    
    # Plot sample trajectories
    initial_conditions = [
        [0.1, 0.8, 0.1], [0.3, 0.3, 0.4], [0.5, 0.1, 0.4],
        [0.6, 0.3, 0.1], [0.2, 0.5, 0.3], [0.4, 0.4, 0.2]
    ]
    for x0 in initial_conditions:
        traj = odeint(replicator_dynamics, x0, t, args=(params, model_name))
        ax.plot(traj[:, 0], traj[:, 1], '-')
    
    # Simplex boundaries
    ax.plot([0, 1], [0, 0], 'k')
    ax.plot([0, 0], [0, 1], 'k')
    ax.plot([0, 1], [1, 0], 'k')
    
    ax.set_xlabel('Proportion of Public (C)')
    ax.set_ylabel('Proportion of Cheater (D)')
    ax.set_title(f'Phase Diagram - {model_name.capitalize()} Model')
    plt.show()

# Plot phase diagrams for both models
plot_phase_diagram(params_first_model, 'first')
plot_phase_diagram(params_second_model, 'second')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the selection intensity range (β_s)
selection_intensity_values = np.linspace(0.1, 5, 50)  # Varying β_s from weak to strong selection
C_values = np.linspace(0, 1, 100)  # Fraction of Cooperators
P_values = np.linspace(0, 1, 100)  # Fraction of Private Metabolizers

# Generate the probability landscape
probability_landscape = np.zeros((len(C_values), len(selection_intensity_values)))

for i, C in enumerate(C_values):
    for j, beta_s in enumerate(selection_intensity_values):
        # Compute fitness differences
        F_C = 0.04 - 0.0001 * C - 0.0002 * (1 - C - P_values[i]) - 0.0001 * P_values[i]
        F_D = 0.0002 * C - 0.00001 * (1 - C - P_values[i])
        F_P = 0.02 - 0.0001 * C - 0.0001 * P_values[i]

        # Compute transition probabilities
        P_C_to_D = 1 / (1 + np.exp(-beta_s * (F_D - F_C)))
        P_C_to_P = 1 / (1 + np.exp(-beta_s * (F_P - F_C)))

        # Compute equilibrium probability of cooperators
        probability_landscape[i, j] = P_C_to_D + P_C_to_P  # Summing probability transitions

# Plot heatmap
plt.figure(figsize=(8, 5))
plt.imshow(probability_landscape, aspect='auto', cmap='coolwarm',
           extent=[selection_intensity_values.min(), selection_intensity_values.max(), C_values.min(), C_values.max()],
           origin='lower')

plt.colorbar(label="Transition Probability")
plt.xlabel("Selection Intensity (β_s)")
plt.ylabel("Fraction of Cooperators (C)")
plt.title("Heatmap of Strategy Transitions for Different Selection Intensities")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameter range for alpha (x-axis) and selection pressure beta_s (y-axis)
alpha_values = np.linspace(0.0001, 0.001, 50)  # Varying alpha (exploitation factor)
beta_s_values = np.linspace(0, 1, 50)  # Selection pressure from weak (0) to strong (1)

# Initialize the heatmap matrix
cooperator_fractions = np.zeros((len(beta_s_values), len(alpha_values)))

# Loop over different values of alpha and beta_s
for i, beta_s in enumerate(beta_s_values):
    for j, alpha in enumerate(alpha_values):
        # Define fixed parameters
        C = 0.5  # Initial cooperator fraction (arbitrary midpoint)
        P = 0.2  # Initial private metabolizer fraction

        # Compute fitness values
        F_C = 0.04 - 0.0001 * C - alpha * (1 - C - P) - 0.0001 * P
        F_D = alpha * C - 0.00001 * (1 - C - P)
        F_P = 0.02 - 0.0001 * C - 0.0001 * P

        # Compute transition probabilities
        P_C_to_D = 1 / (1 + np.exp(-beta_s * (F_D - F_C)))
        P_C_to_P = 1 / (1 + np.exp(-beta_s * (F_P - F_C)))

        # Compute equilibrium fraction of cooperators
        cooperator_fractions[i, j] = 1 - (P_C_to_D + P_C_to_P)

# Plot heatmap
plt.figure(figsize=(8, 5))
plt.imshow(cooperator_fractions, aspect='auto', cmap='coolwarm',
           extent=[alpha_values.min(), alpha_values.max(), beta_s_values.min(), beta_s_values.max()],
           origin='lower')

plt.colorbar(label="Fraction of Cooperators")
plt.xlabel("Exploitation Factor (α)")
plt.ylabel("Selection Pressure (β_s)")
plt.title("Heatmap of Cooperation Fraction vs. Exploitation and Selection Pressure")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameter values
beta = 0.0001  # Competition loss between public and private
gamma = 0.0001  # Cost spent by private population
alpha_default = 0.0002  # Default benefit from cooperator to defector
eta = 0.0001  # Cost spent by public population
r_c = 0.5  # Growth rate of public metabolizers
r_p = 0.2  # Growth rate of private metabolizers
nu = 0.0001  # Transportation cost of defectors

# Define the parameter range for alpha (x-axis) and selection pressure beta_s (y-axis)
alpha_values = np.linspace(0.0001, 0.01, 100)  # Varying exploitation factor (α)
beta_s_values = np.linspace(-10, 10, 100)  # Selection pressure from weak (0) to strong (1)

# Initialize the heatmap matrix for fraction of cooperators
cooperator_fractions = np.zeros((len(beta_s_values), len(alpha_values)))

# Loop over different values of alpha and beta_s
for i, beta_s in enumerate(beta_s_values):
    for j, alpha in enumerate(alpha_values):
        # Define initial strategy fractions
        C_init = 0.5  # Initial fraction of Cooperators
        D_init = 0.3  # Initial fraction of Cheaters
        P_init = 1 - C_init - D_init  # Ensuring sum = 1

        # Compute fitness values based on current strategy distribution
        F_C = r_c - eta * C_init - alpha * D_init - beta * P_init
        F_D = alpha * C_init - nu * D_init
        F_P = r_p - gamma * P_init - beta * C_init

        # Compute transition probabilities based on selection pressure
        P_C_to_D = 1 / (1 + np.exp(-beta_s * (F_D - F_C)))
        P_C_to_P = 1 / (1 + np.exp(-beta_s * (F_P - F_C)))

        # Compute equilibrium fraction of cooperators
        cooperator_fractions[i, j] = C_init * (1 - (P_C_to_D + P_C_to_P))

# Plot heatmap
plt.figure(figsize=(8, 5))
plt.imshow(cooperator_fractions, aspect='auto', cmap='coolwarm',
           extent=[alpha_values.min(), alpha_values.max(), beta_s_values.min(), beta_s_values.max()],
           origin='lower')

plt.colorbar(label="Fraction of public")
plt.xlabel("Exploitation Factor (α)")
plt.ylabel("Selection Pressure (β_s)")
plt.title("Heatmap of Cooperation Fraction vs. Exploitation and Selection Pressure")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the model
r_C = 0.04  
alpha = 0.0002  
beta = 0.0001  
lambda_C = 0.0001  
nu = 0.00001  
r_P = 0.02  
gamma = 0.0001  

# Simulation parameters
time_steps = 100  
C_init, D_init, P_init = 0.5, 0.3, 0.2  

# Arrays to store evolutionary trajectories
C_evolution, D_evolution, P_evolution = [], [], []

# Simulate replicator dynamics
C, D, P = C_init, D_init, P_init
for t in range(time_steps):
    phi = C * (r_C - lambda_C * C - alpha * D - beta * P) + D * (alpha * C - nu * D) + P * (r_P - gamma * P - beta * C)

    dC = C * ((r_C - lambda_C * C - alpha * D - beta * P) - phi)
    dD = D * ((alpha * C - nu * D) - phi)
    dP = P * ((r_P - gamma * P - beta * C) - phi)

    C += dC
    D += dD
    P += dP

    total = C + D + P
    C, D, P = C / total, D / total, P / total

    C_evolution.append(C)
    D_evolution.append(D)
    P_evolution.append(P)

# --- TIME EVOLUTION PLOT ---
plt.figure(figsize=(8, 5))
plt.plot(range(time_steps), C_evolution, label="public (C)", color="green")
plt.plot(range(time_steps), D_evolution, label="cheaters (D)", color="black")
plt.plot(range(time_steps), P_evolution, label="private (P)", color="orange")
plt.xlabel("Time (Generations)")
plt.ylabel("Fraction of Population")
plt.title("Time Evolution of Strategies in Microbial Community")
plt.legend()
plt.grid(True)
plt.show()

# --- SELECTION PRESSURE HEATMAP ---
generations = np.arange(time_steps)
selection_pressure = np.logspace(-2, 2, 50)

heatmap_data = np.zeros((len(selection_pressure), len(generations)))

for i, sp in enumerate(selection_pressure):
    heatmap_data[i, :] = np.exp(-sp * generations) / (1 + np.exp(-sp * generations))

plt.figure(figsize=(8, 5))
plt.imshow(heatmap_data, cmap="coolwarm", aspect="auto", origin="lower",
           extent=[generations[0], generations[-1], selection_pressure[0], selection_pressure[-1]])

plt.colorbar(label="Fraction of Public(C)")
plt.xlabel("Generations")
plt.ylabel("Private selection Pressure ")
plt.title("Effect of Selection Pressure Over Generations")
plt.yscale("log")
plt.show()
```

```python
# --- SELECTION PRESSURE HEATMAP FOR PRIVATE POPULATION (P) ---
heatmap_data_P = np.zeros((len(selection_pressure), len(generations)))

for i, sp in enumerate(selection_pressure):
    heatmap_data_P[i, :] = (1 - np.exp(-sp * generations)) / (1 + np.exp(-sp * generations))

plt.figure(figsize=(8, 5))
plt.imshow(heatmap_data_P, cmap="coolwarm", aspect="auto", origin="lower",
           extent=[generations[0], generations[-1], selection_pressure[0], selection_pressure[-1]])

plt.colorbar(label="Fraction of Private Population (P)")
plt.xlabel("Generations")
plt.ylabel("Selection Pressure")
plt.title(" Selection Pressure on public to private Over Generations")
plt.yscale("log")
plt.show()
```

```python
import matplotlib.pyplot as plt

def draw_decision_tree():
    fig, ax = plt.subplots(figsize=(10, 6))

    # Define nodes and positions
    nodes = {
        "Gen1": (5, 6),
        "Gen2_P_to_C": (3, 4),
        "Gen2_P_to_D": (7, 4),
        "Gen3_P_to_C_C": (2, 2),
        "Gen3_P_to_C_D": (4, 2),
        "Gen3_P_to_D_C": (6, 2),
        "Gen3_P_to_D_D": (8, 2),
    }

    edges = [
        ("Gen1", "Gen2_P_to_C"), ("Gen1", "Gen2_P_to_D"),
        ("Gen2_P_to_C", "Gen3_P_to_C_C"), ("Gen2_P_to_C", "Gen3_P_to_C_D"),
        ("Gen2_P_to_D", "Gen3_P_to_D_C"), ("Gen2_P_to_D", "Gen3_P_to_D_D")
    ]

    labels = {
        "Gen1": "Public (C), Cheater (D), Private (P)",
        "Gen2_P_to_C": "Private → Public",
        "Gen2_P_to_D": "Private → Cheater",
        "Gen3_P_to_C_C": "Public → Public",
        "Gen3_P_to_C_D": "Public → Cheater",
        "Gen3_P_to_D_C": "Cheater → Public",
        "Gen3_P_to_D_D": "Cheater → Cheater",
    }

    # Draw nodes
    for node, (x, y) in nodes.items():
        ax.text(x, y, labels[node], ha='center', va='center',
                bbox=dict(boxstyle="round,pad=0.3", edgecolor="black", facecolor="lightblue"))

    # Draw edges
    for start, end in edges:
        x_values = [nodes[start][0], nodes[end][0]]
        y_values = [nodes[start][1], nodes[end][1]]
        ax.plot(x_values, y_values, 'k-', lw=1.5)

    # Remove axes
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 7)
    ax.axis('off')

    plt.title("Decision Tree of Microbial Evolutionary Transitions")
    plt.show()

# Draw the decision tree
draw_decision_tree()
```
