### Main Goal 2: Use Evolutionary Game Theory to Analyze the Cost and Benefit of Enzyme Production in Heterotrophic Partners

#### Metabolic Strategies

1. **Public Metabolizers**:
    - Produce enzymes to convert sucrose into glucose, which is shared with the community.
    - Incurs a production cost but benefits from the collective pool of glucose.

2. **Private Metabolizers**:
    - Consume sucrose directly without sharing glucose.
    - Avoid enzyme production costs but do not contribute to the communal glucose pool.

3. **Cheaters**:
    - Exploit glucose produced by public metabolizers without producing enzymes.
    - Benefit from public resources without incurring costs, risking communal resource depletion.

#### Analytical Approach

1. **Payoff Matrices**:
    - Quantify the costs and benefits for each strategy based on the frequency of each strategy and resource availability.
    - Example: Public metabolizers balance enzyme production costs with glucose access, while cheaters gain glucose without costs.

2. **Replicator Dynamics**:
    - Model the frequency changes of each strategy over time using replicator equations.
    - Equations: 
      $$
      \frac{d x_i}{dt} = x_i \left( f_i - \bar{f} \right)
      $$
      where $ x_i $ is the strategy frequency, $ f_i $ is the strategy payoff, and $ \bar{f} $ is the average population payoff.

3. **Evolutionary Stable Strategies (ESS)**:
    - Identify conditions where a strategy becomes stable and resistant to invasion by others.
    - Determine the impact of enzyme production costs and resource availability on strategy stability.

4. **Simulation and Numerical Analysis**:
    - Run simulations to observe strategy frequency changes under various conditions.
    - Use numerical methods to solve replicator equations and visualize strategy dynamics.

###  Michaelis-Menten kinetics

Agent behaviors are defined by a set of rules that dictate how agents interact with the environment and other agents. For example:
- **Photosynthesis Rule**: The rate of photosynthesis $(P_i)$ for agent $(i)$ can be modeled as a function of light $(L_i)$ and CO_2 availability $(C_i)$, possibly incorporating Michaelis-Menten kinetics for both factors.
  
  $$ P = f(L_o, C_r) = P_{max} \frac{L_o }{K_L + L_o} \cdot \frac{C_r}{K_C + C_r} $$
  
  where $ K_L $ and $ K_C $ are half-saturation constants for light and CO_2, respectively.

- **Sucrose Production Rule**: The sucrose production $S $ might be a fraction of the photosynthesis output, adjusted by the cell's energy needs and efficiency factors.
  
 
  $$ S(t) = k \cdot P  $$
  where $ k $ is the conversion efficiency of photosynthetic products to sucrose.

## Measuring Optical Density of Cyanobacteria Cultures Using the Beer–Lambert Law

Optical density (OD) measurements are fundamental in quantifying the concentration of cyanobacteria in cultures. By leveraging the Beer–Lambert law, we can relate the absorbance of light by a culture to its cell concentration, providing a robust method for monitoring and analyzing cyanobacterial growth.

###  the Beer–Lambert Law

The Beer–Lambert law, a cornerstone in spectrophotometry, establishes a direct relationship between the absorbance of light by a solution and the concentration of the absorbing substance within that solution. Mathematically, it is expressed as:

$$ A = \epsilon \cdot c \cdot l $$

Where:
- **Absorbance ($A$)**: A unitless measure of light absorbed by the sample.
- **Molar Extinction Coefficient ($ε$)**: A constant that indicates how strongly the substance absorbs light at a particular wavelength, with units L·mol⁻¹·cm⁻¹.
- **Concentration ($c$)**: The concentration of the absorbing substance, typically in mol/L or cells/mL.
- **Path Length ($l$)**: The distance the light travels through the sample, usually in centimeters (cm).

### Applying the Beer–Lambert Law to Cyanobacteria Cultures

When applied to cyanobacteria cultures, the Beer–Lambert law facilitates the quantification of cell concentration based on light absorption. Here’s a step-by-step guide to understanding and implementing this process:

#### Preparation and Measurement

1. **Sample Preparation**: Ensure the cyanobacteria culture is homogeneous by mixing thoroughly before measurement.
2. **Wavelength Selection**: Choose an appropriate wavelength that maximizes absorption by the cyanobacteria. For many cyanobacteria species, a wavelength around 660 nm (OD₆₆₀) is commonly used.
3. **Measurement**: Place the culture sample in a cuvette and use a spectrophotometer to measure the OD at the selected wavelength.

#### The OD Measurement

The optical density (OD) at a specific wavelength is indicative of the culture's turbidity, which correlates with cell concentration. The spectrophotometer provides an OD reading, representing the absorbance of light by the sample.

### Interpreting OD Measurements

The relationship between OD and cell concentration is generally linear at low to moderate concentrations, allowing for straightforward application of the Beer–Lambert law. However, deviations from linearity can occur at higher concentrations due to scattering and other optical effects.

#### Creating a Calibration Curve

To accurately translate OD readings into cell concentrations, a calibration curve is often employed. This curve is generated by plotting known concentrations of cyanobacteria against their corresponding OD readings. Here’s how to construct and use a calibration curve:

1. **Prepare Standard Solutions**: Create a series of cyanobacteria solutions with known concentrations.
2. **Measure OD**: Measure the OD of each standard solution using the spectrophotometer.
3. **Plot the Curve**: Plot the OD values against the known concentrations to generate the calibration curve.

Once the calibration curve is established, it can be used to determine the concentration of unknown samples by matching their OD readings to the curve.

### Example: Calculating Cell Concentration

Suppose you have a cyanobacteria culture and measure its OD₆₆₀ to be 0.6. From previous experiments, you have determined that the molar extinction coefficient (ε) for your cyanobacteria at 660 nm is 1.5 L·mol⁻¹·cm⁻¹. The path length (l) of your cuvette is 1 cm.

Using the Beer–Lambert law:

$$ A = \epsilon \cdot c \cdot l $$

Given:
$$ A = 0.6 $$
$$ \epsilon = 1.5 \text{ L·mol}^{-1}\text{·cm}^{-1} $$
$$ l = 1 \text{ cm} $$

Rearrange the equation to solve for concentration (c):

$$ c = \frac{A}{\epsilon \cdot l} $$

$$ c = \frac{0.6}{1.5 \cdot 1} $$

$$ c = 0.4 \text{ mol/L} $$

If 1 mol of cyanobacteria is equivalent to $ 10^9 $ cells/mL, then:

$$ c = 0.4 \text{ mol/L} \times 10^9 \text{ cells/mol} $$

$$ c = 4 \times 10^8 \text{ cells/mL} $$

Therefore, the concentration of your cyanobacteria culture is $ 4 \times 10^8 $ cells/mL.

### Phototrophic and Heterotrophic Kingdom Dynamics

#### Cyanobacteria Photosynthesis

Cyanobacteria are photosynthetic microorganisms that convert light energy and carbon dioxide (CO₂) into organic carbon in the form of sucrose (Sh) through photosynthesis. The photosynthesis process can be described by Michaelis-Menten kinetics to incorporate the effects of light intensity (L_i) and CO₂ concentration (C_i). The rate of photosynthesis (P_i) for cyanobacteria can be modeled as:

$$  P_i = \frac{P_{\text{max}} \cdot L_i \cdot C_i}{K_L + L_i + \frac{C_i}{K_C}}  $$

where:
- $ P_{\text{max}} $ is the maximum photosynthesis rate,
- $ K_L  $ is the Michaelis constant for light,
- $ K_C  $ is the Michaelis constant for CO₂.

$$ P(t) = k \cdot P_i \cdot C  $$

The produced sucrose (P) is partially consumed by cyanobacteria for growth and maintenance, and the rest is excreted into the environment, making it available for the heterotrophic partners. The sucrose intake rate equation is:

$$ S(t) = L_0 \left( \frac{S_f}{V} \right) \epsilon\cdot C  $$

with:
- $ I_0 $  being the intensity of incident light,
- $ S_f $ being the illuminated surface area of the reactor,
- $ V $ being the reactor's liquid volume,
- $ C $ total cyano biomass,
- $ ε $  Extinction Coefficient 
- $ t $ being the total time of production,


$$ S_h =  P(t) - S(t) $$

where $ S_h $ is the sucrose concentration rate given to the heterotropic partner. 

The growth rate of cyanobacteria $ \mu_C $ is dependent on the available sucrose:

$ \mu_C = \mu_{\text{max}} \frac{S(t)}{K_s + S(t)} $

and their biomass change over time is given by:

$ \frac{dC}{dt} = \mu_C \cdot C - \lambda \cdot C $

where:
- $ \mu_{\text{max}} $ is the maximum specific growth rate,
- $ K_s $ is the half-saturation constant for growth,
- $ \lambda $ is the mortality rate.

#### Heterotrophic Partners

The heterotrophic partners in the community utilize the excreted sucrose from cyanobacteria through different metabolic strategies. These partners are divided into two groups: 1) Public metabolizer and cheater, 2) Private metabolizer and public metabolizer.

##### 1) Public Metabolizer and Cheater

- **Public Metabolizer (P):** This partner produces public enzymes that convert sucrose (Sh) into glucose (G). The specific growth rate $ \mu_P $ of the public metabolizer depends on the available sucrose:

  $$ \mu_P = \mu_{\text{max}} \frac{G}{K_p + G} $$

  Their biomass change over time is given by:

  $$ \frac{dP}{dt} = \mu_P \cdot P - \lambda \cdot P - \beta \cdot P $$

  where $ \beta $ represents the cost of producing public enzymes.

- **Cheater (D):** This partner does not produce public enzymes and relies on the glucose (G) produced by the public metabolizer. The specific growth rate $ \mu_D $ of the cheater is:

  $$ \mu_D = \mu_{\text{max}} \frac{G}{K_D + G} $$

  Their biomass change over time is given by:

  $$ \frac{dD}{dt} = \mu_D \cdot D - \lambda \cdot D $$

The change in glucose concentration (G) over time is described by:

$$ \frac{dG}{dt} = e_P \cdot S_h * P- r_p \cdot P  -  r_c \cdot D $$

$e_P$ is the enzyme producer sucrose to glucose production rate. $ r_p$ and $ r_c$ are the consumption rate.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Constants
K_L = 200  # Michaelis constant for light (μmol photons m⁻² s⁻¹)
K_C = 300  # Michaelis constant for CO2 (ppm)
alpha = 0.8  # Conversion factor from photosynthesis rate to sucrose production rate
P_max = 0.00000000013  # Maximal net rate of photosynthesis (μmol O2 s⁻¹ mg chl a⁻¹)
I_0 = 1100  # Intensity of incident light (μmol photons m⁻² s⁻¹)
S_f = 1.0  # Illuminated surface area of the reactor (m²)
V = 0.01  # Reactor's liquid volume (L)
C_init = 1000  # Initial cyanobacteria biomass concentration (mg/L)
epsilon = 0.0001  # Extinction Coefficient
mu_max = 0.005   # Maximum specific growth rate (1/s)
K_s = 0.1  # Half-saturation constant for growth (μmol sucrose/L)
lambda_ = 0.004   # Mortality rate (1/s)
beta = 0.00005  # Cost of producing public enzymes (1/s)
h_P = 0.001  # Conversion rate of sucrose to glucose by public metabolizers (1/s)
K_p = 0.1  # Half-saturation constant for public metabolizer growth (μmol glucose/L)
C_i = 500  # Fixed CO2 concentration (ppm)
P_init = 10  # Initial public metabolizer biomass concentration (mg/L)
D_init = 1  # Initial cheater biomass concentration (mg/L)
G_init = 0  # Initial glucose concentration (μmol/L)

# Time range (hours, 250 hours)
time_range = np.linspace(0, 160, 1000)

# Define the photosynthesis function using Michaelis-Menten kinetics
def photosynthesis_rate(L_i, C_i, K_L, K_C, P_max):
    return (P_max * L_i * C_i) / (K_L + L_i + (C_i / K_C))

# Define the total sucrose production function
def total_sucrose_production(P_i, alpha, C):
    return alpha * P_i * C

# Define the sucrose intake rate function
def sucrose_intake_rate(I_0, S_f, V, epsilon, C):
    return I_0 * (S_f / V) * epsilon * C

# Define the growth rate function
def growth_rate(S, mu_max, K_s):
    return mu_max * (S / (K_s + S))

# ODE function for biomass and glucose change over time
def system(state, t, L_i, C_i, K_L, K_C, P_max, alpha, S_f, V, epsilon, mu_max, K_s, K_p, lambda_, beta, h_P):
    C, P, D, G = state

    # Photosynthesis and sucrose production
    P_i = photosynthesis_rate(L_i, C_i, K_L, K_C, P_max)
    P_sucrose = total_sucrose_production(P_i, alpha, C)
    S = sucrose_intake_rate(I_0, S_f, V, epsilon, C)
    S_h = P_sucrose - S

    # Public Metabolizer growth rate and biomass change
    mu_P = mu_max * (G / (K_p + G))
    dPdt = mu_P * P - lambda_ * P - beta * P

    # Cheater growth rate and biomass change
    mu_D = mu_max * (G / (K_s + G))
    dDdt = mu_D * D - lambda_ * D

    # Glucose dynamics
    dGdt = h_P * S_h * P - mu_P * P - mu_D * D
    
    # Cyanobacteria growth rate and biomass change
    mu_C = growth_rate(S, mu_max, K_s)
    dCdt = mu_C * C - lambda_ * C

    return [dCdt, dPdt, dDdt, dGdt]

# Initial state
state_init = [C_init, P_init, D_init, G_init]

# Solve ODE for system dynamics over time
solution = odeint(system, state_init, time_range * 3600, args=(I_0, C_i, K_L, K_C, P_max, alpha, S_f, V, epsilon, mu_max, K_s, K_p, lambda_, beta, h_P))
C_values, P_values, D_values, G_values = solution.T

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(time_range, C_values, label='Cyanobacteria Biomass ($C$)')
plt.plot(time_range, P_values, label='Public Metabolizer Biomass ($P$)')
plt.plot(time_range, D_values, label='Cheater Biomass ($D$)')
#plt.plot(time_range, G_values, label='Glucose ($G$)')

plt.xlabel('Time (hours)')
plt.ylabel('Concentration')
plt.title('System Dynamics over Time')
plt.legend()
plt.grid(True)
plt.show()
```

##### 2) Private Metabolizer and Public Metabolizer

- **Private Metabolizer (Q):** This partner produces private enzymes to convert sucrose (Sh) into glucose (G) for its own use. The specific growth rate $ \mu_P $ is:

  $$ \mu_q = \mu_{\text{max}} \frac{S_q}{K_s + S_q} $$

  Their biomass change over time is given by:

  $$ \frac{dQ}{dt} = \mu_q \cdot Q - \lambda \cdot Q - \alpha \cdot Q $$
  
  where $ \alpha $ represents the cost of producing public enzymes.

- **Public Metabolizer (P):** As described above, this partner produces public enzymes and converts sucrose into glucose for itself and others.
$$  S_p = Sh - S_q   $$
$$ G = h_P \cdot S_p \cdot P $$


$$ \mu_p = \mu_{\text{max}} \frac{G}{K_g + G} $$

$$ \frac{dP}{dt} = \mu_P \cdot P - \lambda \cdot P - \beta \cdot P $$
      
The change in glucose concentration (G) and sucrose concentration (Sh) over time is described by:

$$ \frac{dG}{dt} = h_P \cdot S_p \cdot P - \mu_P \cdot P- common pool glucose $$

### Integrated Model

The combined dynamics of cyanobacteria and heterotrophic partners are described by the following system of differential equations:

1. **Cyanobacteria (C):**
   $$ \frac{dC}{dt} = \mu_C \cdot C - \lambda \cdot C $$

2. **Cheater (D):**
   $$ \frac{dD}{dt} = \mu_D \cdot D - \lambda \cdot D $$

3. **Public Metabolizer (P):**
   $$ \frac{dP}{dt} = \mu_P \cdot P - \lambda \cdot P - \beta \cdot P $$

4. **Sucrose (Sh):**
   $$ S_h = P_i - S - h_P \cdot Sh \cdot P \ $$

5. **Glucose (G):**
   $$ \frac{dG}{dt} = h_P \cdot Sh \cdot P - \mu_P \cdot \left( \frac{G}{K_s + G} \right) \cdot D $$

These equations model the interplay between the phototrophic cyanobacteria and their heterotrophic partners, capturing the essential dynamics of resource production, consumption, and competition. The cyanobacteria produce sucrose through photosynthesis, which is then utilized by heterotrophic partners either publicly or privately, shaping the ecosystem's structure and function.

```python
import numpy as np
from scipy.optimize import fsolve
from numpy.linalg import eigvals

# Parameters
P_max = 2.0        # Maximum photosynthesis rate (µmol O2 min⁻¹ mg chl a⁻¹)
K_L = 100.0        # Michaelis constant for light (µmol m⁻² s⁻¹)
K_C = 0.1          # Michaelis constant for CO2 (mmol L⁻¹)
I_0 = 500.0        # Intensity of incident light (µmol m⁻² s⁻¹)
S_f = 0.05         # Illuminated surface area (m²)
V = 1.0            # Reactor liquid volume (L)
A = 0.1            # Absorbance (OD)
l = 1.0            # Path length of the cuvette (cm)
mu_max = 1.0       # Maximum specific growth rate (day⁻¹)
K_s = 0.05         # Half-saturation constant (mmol L⁻¹)
lambda_ = 0.01     # Mortality rate (day⁻¹)
beta = 0.02        # Cost of producing public enzymes (day⁻¹)
h_P = 0.1          # Public enzyme conversion efficiency (mmol sucrose mg⁻¹ day⁻¹)

# Define the system of equations
def equations(vars):
    C, Pri, Pub, Sh, G = vars

    L_i = I_0 * (S_f / V) * (A / (C * l))
    P_i = (P_max * L_i * C) / (K_L + L_i + (C / K_C))

    mu_C = mu_max * Sh / (K_s + Sh)
    mu_Pri = mu_max * Sh / (K_s + Sh)
    mu_Pub = mu_max * Sh / (K_s + Sh)

    eq1 = mu_C * C - lambda_ * C
    eq2 = P_i - h_P * Sh * Pub - Pri * Sh
    eq3 = h_P * Sh * Pub + Pri * Sh - mu_max * (G / (K_s + G)) * (Pub + Pri)
    eq4 = mu_Pri * Pri - lambda_ * Pri
    eq5 = mu_Pub * Pub - lambda_ * Pub - beta * Pub

    return [eq1, eq2, eq3, eq4, eq5]

# Initial guess for the variables
initial_guess = [0.5, 0.1, 0.1, 1.0, 0.2]

# Solve the system of equations
solution = fsolve(equations, initial_guess)
C, Pri, Pub, Sh, G = solution

print("Steady-State Solutions:")
print(f"Cyanobacteria (C): {C:.4f}")
print(f"Private Metabolizer (Pri): {Pri:.4f}")
print(f"Public Metabolizer (Pub): {Pub:.4f}")
print(f"Sucrose (Sh): {Sh:.4f}")
print(f"Glucose (G): {G:.4f}")

# Define the Jacobian matrix
def jacobian(vars):
    C, Pri, Pub, Sh, G = vars

    L_i = I_0 * (S_f / V) * (A / (C * l))
    P_i = (P_max * L_i * C) / (K_L + L_i + (C / K_C))

    mu_C = mu_max * Sh / (K_s + Sh)
    mu_Pri = mu_max * Sh / (K_s + Sh)
    mu_Pub = mu_max * Sh / (K_s + Sh)

    dL_dC = -I_0 * (S_f / V) * (A / (C**2 * l))

    dP_dC = (P_max * L_i) / (K_L + L_i + (C / K_C)) + \
            (P_max * L_i * C * (-dL_dC / (K_L + L_i + (C / K_C)))) + \
            (P_max * L_i * C * (-C / (K_C * (K_L + L_i + (C / K_C))**2)))

    dP_dSh = 0
    dP_dPri = 0
    dP_dPub = 0
    dP_dG = 0

    J = np.array([
        [mu_C - lambda_, 0, 0, C * mu_max * K_s / (K_s + Sh)**2, 0],
        [dP_dC, -Sh, -h_P * Sh, -h_P * Pub - Pri, 0],
        [0, Sh, h_P * Sh, h_P * Pub + Pri - mu_max * K_s * (Pub + Pri) / (K_s + G)**2, mu_max * (Pub + Pri) * K_s / (K_s + G)**2],
        [0, mu_Pri, 0, Pri * mu_max * K_s / (K_s + Sh)**2, 0],
        [0, 0, mu_Pub - beta, Pub * mu_max * K_s / (K_s + Sh)**2, 0]
    ])
    
    return J

# Calculate the Jacobian matrix at the steady-state solution
J = jacobian(solution)

# Calculate the eigenvalues of the Jacobian matrix
eigenvalues = eigvals(J)

print("\nEigenvalues of the Jacobian Matrix:")
for i, val in enumerate(eigenvalues):
    print(f"Eigenvalue {i+1}: {val:.4f}")

# Check the stability
stable = all(np.real(eigenvalue) < 0 for eigenvalue in eigenvalues)
if stable:
    print("\nThe system is stable at the steady-state solution.")
else:
    print("\nThe system is unstable at the steady-state solution.")
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define the system of differential equations
def deriv(y, t, nu, beta, beta1, alpha, mu, gamma, eta):
    C, D, P = y
    dCdt = mu * C - alpha * C * D - beta * C * P - beta1 * C * C  # Cooperator population
    dDdt = -nu * D * D + alpha * C * D  # Defector population
    dPdt = eta * P - beta * C * P - gamma * P * P  # Private population
    return dCdt, dDdt, dPdt

# Parameters
beta = 0.00001  # Benefit from cooperator to private
beta1 = 0.000001  # Coefficient for cooperator self-impact
gamma = 0.00001  # Cost spent by private population
alpha = 0.00001  # Benefit from cooperator to defector
mu = 0.009  # Growth rate of cooperator
eta = 0.005  # Growth rate of private
nu = 0.000001  # Transportation cost of defector

# Initial conditions
C0, D0, P0 = 10, 1, 1
y0 = C0, D0, P0

# Time grid
t = np.linspace(0, 1000, 1000)  # Total time of 600 seconds

# Integrate the equations over the time grid
ret = odeint(deriv, y0, t, args=(nu, beta,beta1, alpha, mu, gamma, eta))

# Plot the results
C, D, P = ret.T

# Function to plot populations over time
def plot_populations(t, C, D, P):
    plt.figure(figsize=(10, 6))
    plt.plot(t, C, 'b', label='Public Population (C)')
    plt.plot(t, D, 'g', label='Cheater Population (D)')
    plt.plot(t, P, 'orange', label='Private Population (P)')
    plt.xlabel('Time (seconds)')
    plt.ylabel('Population')
    plt.title('Population Dynamics over Time')
    plt.legend()
    plt.grid()
    plt.show()

# Phase portraits
def plot_phase_portraits(C, D, P):
    plt.figure(figsize=(10, 6))
    plt.subplot(1, 2, 1)
    plt.plot(D, C)
    plt.xlabel('Defector Population (D)')
    plt.ylabel('Cooperator Population (C)')
    plt.title('Phase Portrait: C vs. D')
    
    plt.subplot(1, 2, 2)
    plt.plot(C, P)
    plt.xlabel('Public Population (C)')
    plt.ylabel('Private Population (P)')
    plt.title('Phase Portrait: C vs. P')
    
    plt.tight_layout()
    plt.show()

# Call the plotting functions
plot_populations(t, C, D, P)
plot_phase_portraits(C, D, P)
```
